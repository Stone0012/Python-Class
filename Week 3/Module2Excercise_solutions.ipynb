# Logistic Regression

<font size =4>Let's do a logistic regression for the [digits data set](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html).  (smaller than MNIST)

```python
from sklearn import datasets
digits = datasets.load_digits()
```

```python
X, y = digits["data"], digits["target"]
X.shape, np.min(X),np.max(X)
```

<font size=4 color =blue>Do a min max scaling.  

```python
X=X/16
```

##  Train test split

<font size=4 color=blue> Split the data using train_test_split to do a create a test set with 15% of the data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

<font size=4 color =blue> Fit a Logistic regression to the digits data set.  Use  
    
    max_iter=5000,random_state=42,C=1

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
log_reg = LogisticRegression(max_iter=5000,random_state=42,C=1)
log_reg.fit(X_train, y_train)
scores=cross_val_score(log_reg, X_train, y_train, cv=3, scoring="accuracy")
print("scores", scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

<font size=4 color =blue> See how it performs on the test set.

```python
log_reg.score(X_test,y_test)
```

### Bonus Grid Search

<font size =4> Let's do a deeper dive on the [Logisitc Regression parameters](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) with [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)

```python
from sklearn.model_selection import GridSearchCV
log_grid=LogisticRegression(max_iter=5000)
params=[{'C': [.001,.1,1,100],'solver':['lbfgs', 'sag']}]
clf = GridSearchCV(log_grid,params,cv=4,verbose = 0, scoring = 'accuracy') #verbose supresses some of the prgress output 
clf.fit(X_train,y_train)
print("best params",clf.best_params_,"bestscore",clf.best_score_)

```

<font size=4> Let's store the results in a data frame so they are easier to read

```python
import pandas as pd
gridresults=pd.DataFrame(clf.cv_results_)
gridresults
```

<font size=4> That is a a lot of output so let's just select a few of the columns. 

```python
DF=gridresults.sort_values(by=['rank_test_score'])
DF[['param_C','param_solver', 'mean_test_score',"rank_test_score"]]
```

<font size=4> Finally let's run the best model on the Test set we put aside.

```python
clf.best_estimator_.score(X_test,y_test)
```

# SGD Classifier for Iris Data set.  All Features all classes.

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
X=iris.data
y=iris.target
X.shape,y.shape
```

<font size=4 color =blue> Load the data into variables $X$ and $y$.

<font size=4 color=blue> Let's scale the data this time using a standard scaler. Beacuse this data set is so small we will not split off a test set.

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()  
Xs=SS.fit_transform(X)
```

<font size=4 color =blue> Do a Cross validation with cv=4. Use the following parameters for SGD
    
        SGDClassifier(random_state=42)

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(Xs, y)
from sklearn.model_selection import cross_val_score
scores=cross_val_score(sgd_clf, Xs, y, cv=4, scoring="accuracy")
print("scores", scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

<font size=4 color=blue> Create a confusion matrix for your model.

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
y_pred = cross_val_predict(sgd_clf, Xs, y, cv=4)
conf_mx=confusion_matrix(y, y_pred)
print(pd.DataFrame(conf_mx))
```

## Grid Search: 

<font size=4 color =blue> Use example above to try a Grid search for SGD on the Iris data set using 

    params=[{'penalty':['l2', 'l1', 'elasticnet'], 'loss':['hinge', 'log_loss', 'perceptron']}]

[SGD parameters](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html)
    
What is the best model and the paremaetrs of the best model?  Display in a data frame.


```python
from sklearn.model_selection import GridSearchCV
sgd_grid = SGDClassifier(random_state=42,max_iter=500)
params=[{'penalty':['l2', 'l1', 'elasticnet'], 'loss':['hinge', 'log_loss', 'perceptron']}]
clf_grid = GridSearchCV(sgd_grid,params,cv=4,verbose = 1, scoring = 'accuracy') #verbose supresses some of the prgress output 
clf_grid.fit(Xs,y)
print("best C",clf_grid.best_params_,"bestscore",clf_grid.best_score_)
gridresults=pd.DataFrame(clf_grid.cv_results_)
```

```python
DF=gridresults.sort_values(by=['rank_test_score'])
DF[['param_loss','param_penalty', 'mean_test_score']]
```
