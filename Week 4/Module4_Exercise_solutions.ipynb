```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
from numpy import linalg as LA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 
from sklearn.datasets import fetch_openml
```

<font Size=4 Color=blue> Load the digits data set then:
    
    1) Train test Split
    2) StandardScale Fit Transform the Train set.  Transform the Test set.
    3) Find the shape of the scaled Train set  

```python
from sklearn.datasets import load_digits
X,y = load_digits(return_X_y=True)
print(X.shape)
```

```python
SS=StandardScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.15)
```

```python
X_trainS=SS.fit_transform(X_train)
X_testS=SS.transform(X_test) 
X_trainS.shape
```

<font size=4 color=blue>  Apply a Logistic Regression to scaled Train set. 

```python
from sklearn.linear_model import LogisticRegression
log_clf = LogisticRegression()
log_clf.fit(X_trainS, y_train)
log_clf.score(X_trainS,y_train)
```

<font size=4 color=blue> Try Model on the Scaled test set.

```python
log_clf.score(X_testS,y_test)
```

<font size=4 color=blue>  Now do a PCA keeping 90% of the variance for the scaled train set.  What is the new shape of the scaled Train set set?

```python
from sklearn.decomposition  import PCA
pca=PCA()
```

```python
pca.fit(X_trainS)
pca=PCA(n_components=.9)
X_trainS_reduced=pca.fit_transform(X_trainS)
X_trainS_reduced.shape
```

<font size=4 color =blue> Perform the logistic regression for the reduced data. 

```python
log_clf = LogisticRegression(max_iter=115)
#log_clf = LogisticRegression()
log_clf.fit(X_trainS_reduced, y_train)
log_clf.score(X_trainS_reduced,y_train)
```

<font size=4 color=blue> Try Model on the Scaled test set.

```python
X_testS_transformed=pca.transform(X_testS)
log_clf.score(X_testS_transformed,y_test)
```
