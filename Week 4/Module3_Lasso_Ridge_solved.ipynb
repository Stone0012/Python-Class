```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
reg=linear_model.LinearRegression()
import pandas as pd
np.random.seed(42)
```

<font size=4> Load auto data.

```python
autompg=pd.read_fwf("https://raw.githubusercontent.com/lammersm/Data/main/atuoMpg.txt",
                    header=None,names=["mpg","cylinders","discplacment","horespower","weight", "acceleration","modelyear","origin"])
autompg.head()
```

<font size=4 color =blue>  Turn mpg into target and rest of variables into feature set.

```python
auto=np.array(autompg)
ya=auto[:,0]
xa=auto[:,1:8]
```

<font size=4 color =blue> Perform a Ridge Regression
 
* Scale the feature set.    
* Do a GridSearchCV with  $\alpha \in$ [0,1,3,6,9,12] with 4 folds. 
* Display results in a data Frame
* What is the best model?
* Is it better than the model above with no regularization?


```python
from sklearn.linear_model import Ridge
import pandas as pd

from sklearn.preprocessing import StandardScaler
SS =StandardScaler()
xa=SS.fit_transform(xa)

from sklearn.model_selection import GridSearchCV

params={'alpha': [0,1,3,6,9,12]}
rdg_reg = Ridge()
clf = GridSearchCV(rdg_reg,params,cv=4,verbose = 0)

clf.fit(xa,ya)
print("best Alpha",clf.best_params_,"bestscore",clf.best_score_)

DF=pd.DataFrame(clf.cv_results_)
#DF
DF[["param_alpha","mean_test_score","rank_test_score"]]

```

<font size=4 color =blue> __What is the best model ?__

```python
clf.best_estimator_
```

<font size=4 color =blue> Score it on all the data.

```python
clf.best_estimator_.score(xa,ya)
```

<font size=4 color =blue> Perform a Lasso Regression
 
* Scale the feature set.    
* Do a GridSearchCV with  $\alpha \in$ [.1,1,3,6,9,12] with 4 folds. Fine tune the model. 
* Display results in a data Frame
* What is the best model?
* Is it better than the model above with no regularization?

```python
from sklearn.linear_model import Lasso

params={'alpha': [0.1,1,3,6,9,12]}
#params={'alpha': np.linspace(.01,.1,10)}
lass_reg = Lasso()
clf = GridSearchCV(lass_reg,params,cv=4,verbose = 0)

clf.fit(xa,ya)
print("best Alpha",clf.best_params_,"bestscore",clf.best_score_)


DF=pd.DataFrame(clf.cv_results_)
#DF
DF[["param_alpha","mean_test_score","rank_test_score"]]
```

<font size=4 color =blue>__Score the best model on the whole data set?__

```python
print('Best model from grid search can be foud and evaluated on all the data automatically')
clf.best_estimator_.score(xa,ya)
```

<font size=4 color =blue>  What is the best model?

```python
print('What is the best model')
clf.best_estimator_
```
