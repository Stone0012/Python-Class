# Logistic Regression  For Titanic data set

<font size=4 >   See [here](https://towardsdatascience.com/a-beginners-guide-to-kaggle-s-titanic-problem-3193cb56f6ca) for more on the Titanic data set. We are interested in predicting if a passenger survived.
    

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/lammersm/Data/main/titanic.csv")
df
```

<font size=4 color=blue > There are a lot of NANs (missing values) for this data set.  Let's count them by feature.

```python
df.isna().sum()
```

<font size=4 color=blue> Find the correlation matrix.  Notice that correlation only does the numeric values. 

```python
df.corr()
```

<font  size=4> Let's peel off some of the features in a new data frame and drop the Nans.  

```python
df_sub=df[["Survived","Fare",'Sex', "Pclass","Embarked"]]
print(df_sub.shape)
df_sub=df_sub.dropna()
print(df_sub.shape)
```

<font size=4 color=blue> Separate the categorical, the numeric and the target.( I did all the analysis with data frames and did not have to switch to np arrays)  

```python
df_num=(df_sub[["Fare"]])
df_cat=(df_sub[['Sex', "Pclass","Embarked"]])
df_surv=df_sub["Survived"] 
```

<font size=4 color=blue > Logistic regression requires numeric values so create a model using only fare.  use parameter 
    
    C=10 and random_state=42

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression( C=10,random_state=42)# The C value 
log_reg.fit(df_num, df_surv)
```

<font size=4  color =blue> Do a 5 fold CV and print the average of the scores.

```python
from sklearn.model_selection import cross_val_score
import numpy as np
cvs=cross_val_score(log_reg, df_num, df_surv, cv=5
                    , scoring='accuracy')
print(cvs)
print("Average of the scores",np.mean(cvs))
```

# Using  Categorical data as a feature for classification.

<font size=4 color =blue> Apply a one hot encoder to categorical data. Compare original categorical with and one-hot encoded data.

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(drop='if_binary',sparse_output=False).set_output(transform="pandas")
hot_str=enc.fit_transform(df_cat)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression( C=10,random_state=42)# The C value 
log_reg.fit(X_finalS, df_surv)
```
