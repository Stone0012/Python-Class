```python
covid = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv") 
covid.head()
```

<font size =4?> We want to turn this into time series and group by state. It currently has County level information.  We do this with the pandas attribute [groupby](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html?highlight=groupby#pandas.DataFrame.groupby).   When we do this it Turns Province_State into and index for the data frame instead of a feature.  above the index was just a numbering.  Note no label on first column above.

```python
csum=covid.groupby(['Province_State']).sum()
csum.head(10)
```

<font size=4> We want to throw everything away except the last 8 columns containing the data for the last 8 days. This is a cumulative total of the deaths for each state in the lat eight days.

```python
csumT=csum.iloc[:,-8:]
csumT.head()
```

<font size =4>  Let's add a feature.  Number of fatalities in the last 7 days. 

```python
csumT["diff"]=csumT.iloc[:,7]-csumT.iloc[:,0]
csumT.head()
```

<font size=4> Let's rename some features and drop some of the data we don't need.

```python
st_date="3/2/23"
end_date="3/9/23"
csumT1=csumT.rename(columns={"Province_State":"State",st_date:"Start",end_date:"Finish"})
csumT1.drop(csumT1.columns[[1,2,3,4,5,6]],axis=1,inplace=True) 
csumT1.head()
```

# Now let's add another data Set with state populations for each from my github 

<font size=4> I got it from https://worldpopulationreview.com/states
    

https://raw.githubusercontent.com/lammersm/Data/90d67de27440edeac9b3f0e23319e8742398c024/StatePoulation.csv

<font size=4>  Above we calculated the seven day increase in fatalities by state.    However, we would expect states with much larger populations would have larger number of fatalities. So lets combine these data sets and compute two things.
    
1) Total Covid Fatalities per population of the state.  We will  call that 
    
    Deaths Percent of Population
    
2) New Fatalities Per 100K people.  We will call the 
    
    NewDeaths Per 100K


```python
uspop=pd.read_csv("https://raw.githubusercontent.com/lammersm/Data/90d67de27440edeac9b3f0e23319e8742398c024/StatePoulation.csv") 
uspop.head(10)
```

```python
newtopstate=csumT1.merge(uspop,left_on="Province_State",right_on="State")
newtopstate["Deaths Percent of Population"]=100*(newtopstate["Finish"]/newtopstate["Pop"])
newtopstate["NewDeaths Per 100K"]=100000*(newtopstate["diff"]/newtopstate["Pop"])
newtopstate.head()

#newtopstate.to_csv("statetopp_deaths.csv")
```

<font size=4>Let's drop more data that we don't want in our final data set.

```python
newtopstate.drop(newtopstate.columns[[3,6,7,8,9,10,11]],axis=1,inplace=True) 
newtopstate.head()
```

<font size=4> __Let's find out what state has the most New Covid fatalities per 100K people in the last 7 days.__ 

```python
newtopstate=newtopstate.sort_values(by=["NewDeaths Per 100K"],ascending =False)
newtopstate.head(10)
```

<font size=4 color =red> __Find out what state has the most Covid Fatalities  by population__  

```python
newtopstate=newtopstate.sort_values(by=["Deaths Percent of Population"],ascending =False)
newtopstate.head(10)
```

<font size =4> Finally let's write this to a csv file that we can use in Power BI or Tableau. <font color=blue> Check the folder this notebook is in after you run the code below.
